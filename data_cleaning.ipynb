```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import duckdb as db
from datasets import load_dataset
from dotenv import load_dotenv
```

```python
%matplotlib inline
```

```python
# 1. Load the hidden variables from your .env file
load_dotenv()

# 2. Grab the token securely
my_token = os.getenv("HF_TOKEN")

# 3. Pass the token into the load_dataset function
dataset = load_dataset(
    "foursquare/fsq-os-places",
    split="train",
    token=my_token
)
```

```python
# Export the entire dataset straight to a local Parquet file
dataset.to_parquet("all_foursquare_locations.parquet")
```

```python
# Query the entire global dataset
sql_query = """
    SELECT COUNT(*) AS num_rows 
    FROM 'all_foursquare_locations.parquet';
"""

results = db.query(sql_query).df()
display(results)
```

```python
# Query the entire global dataset
sql_query = """
    SELECT * 
    FROM 'all_foursquare_locations.parquet'
    WHERE country = 'AU';
"""

results = db.query(sql_query).df()
```

```python
# Save the filtered Australian locations to a compressed Parquet file
results.to_parquet('au_locations.parquet', index=False)
```

```python
# Display all states in AUS
sql_query = """
    SELECT name, latitude, longitude, address, locality, region, date_closed, tel, website, email, fsq_category_labels
    FROM 'au_locations.parquet'
"""

aus_results = db.query(sql_query).df()
display(aus_results)
```

```python
# Display CLEANED set via state
state = input("State ('all' for all states): ").lower()

if state == "all":
    sql_query = """
        SELECT name, latitude, longitude, address, locality, region, date_closed, tel, website, email, fsq_category_labels
        FROM 'au_locations.parquet'
        WHERE latitude IS NOT NULL AND longitude IS NOT NULL
    """
else:
    sql_query = f"""
        SELECT name, latitude, longitude, address, locality, region, date_closed, tel, website, email, fsq_category_labels
        FROM 'au_locations.parquet'
        WHERE LOWER(region) = '{state}' AND latitude IS NOT NULL AND longitude IS NOT NULL
    """

aus_results = db.query(sql_query).df()
display(aus_results)
```
